```python
df = pd.read_csv('/content/gender_submission.csv')

print("First 5 Rows:")
print(df.head())

print("\nDataset Shape:")
print(df.shape)

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
X = df[['PassengerId']]
y = df['Survived']

print("Features:")
print(X.head())

print("Target:")
print(y.head())
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

print("Training Shape:", X_train.shape)
print("Testing Shape:", X_test.shape)
```

```python
model = LogisticRegression(max_iter=1000)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

base_accuracy = accuracy_score(y_test, y_pred)

print("Base Model Accuracy:", base_accuracy)
```

```python
sfs_forward = SFS(
    model,
    k_features=1,
    forward=True,
    floating=False,
    scoring='accuracy',
    cv=5
)

sfs_forward.fit(X_train, y_train)

selected_forward_features = list(sfs_forward.k_feature_idx_)

print("Selected Forward Features:")
print(X.columns[selected_forward_features])
```

```python
X_train_sfs = X_train[:, selected_forward_features]
X_test_sfs = X_test[:, selected_forward_features]

model.fit(X_train_sfs, y_train)

y_pred_sfs = model.predict(X_test_sfs)

forward_accuracy = accuracy_score(y_test, y_pred_sfs)

print("Forward Selection Accuracy:", forward_accuracy)
```

```python
sfs_backward = SFS(
    model,
    k_features=1,
    forward=False,
    floating=False,
    scoring='accuracy',
    cv=5
)

sfs_backward.fit(X_train, y_train)

selected_backward_features = list(sfs_backward.k_feature_idx_)

print("Selected Backward Features:")
print(X.columns[selected_backward_features])
```

```python
X_train_sbe = X_train[:, selected_backward_features]
X_test_sbe = X_test[:, selected_backward_features]

model.fit(X_train_sbe, y_train)

y_pred_sbe = model.predict(X_test_sbe)

backward_accuracy = accuracy_score(y_test, y_pred_sbe)

print("Backward Elimination Accuracy:", backward_accuracy)
```

```python
print("\n================ PERFORMANCE COMPARISON ================\n")

print("Base Logistic Regression Accuracy:", base_accuracy)

print("Sequential Forward Selection Accuracy:", forward_accuracy)

print("Sequential Backward Elimination Accuracy:", backward_accuracy)
```

```python
results = pd.DataFrame({
    'Method': ['Base Model', 'Forward Selection', 'Backward Elimination'],
    'Accuracy': [
        float(base_accuracy),
        float(forward_accuracy),
        float(backward_accuracy)
    ]
})

print("\n================ FINAL RESULTS ================\n")
print(results.to_string(index=False))
```
